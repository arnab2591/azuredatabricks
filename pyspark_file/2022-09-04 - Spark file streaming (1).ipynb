```
sch = "ID int,Name String,Age int"
df = spark.readStream.option("header",True).schema(sch).csv('/FileStore/tables/student_database/')
display(df)
```

```
# append,complete and update modes of writestream command
df.writeStream.format('console').outputMode('append').start()
```

```
sch = "ID int,Name String,Age int"
df = spark.readStream.option("header",True).schema(sch).option("maxfilespertrigger",1).csv('/FileStore/tables/student_database/')
display(df)
```

```
df.writeStream.format('console').trigger(processingTime="10 seconds").outputMode('append').start()
```

```
sch = "ID int,Name String,Age int"
df = spark.readStream.option("header",True).schema(sch).option("maxfilespertrigger",1).csv('/FileStore/tables/student1/')
display(df)
```

```
df.writeStream.format('memory').queryName("streaming_data").trigger(processingTime="5 seconds").outputMode('append').start()
```

```
%sql
select * from streaming_data
```

```
sch = "ID int,Name String,Age int"
df = spark.readStream.option("header",True).schema(sch).option("maxfilespertrigger",1).csv('/FileStore/tables/student2/')
display(df)
```

```
df1 = df.select("*").groupBy('ID').count() 
```

```
df1.writeStream.format('memory').queryName("streaming_data1").trigger(processingTime="5 seconds").outputMode('complete').start()
```

```
%sql
select * from streaming_data1
```
