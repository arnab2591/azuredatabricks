```
sch = "ID int,Name String,Age int"
df = spark.readStream.option("header",True).schema(sch).csv('/FileStore/tables/student_database/')
display(df)
```

```
# append,complete and update modes of writestream command
df.writeStream.format('console').outputMode('append').start()
```

```
sch = "ID int,Name String,Age int"
df = spark.readStream.option("header",True).schema(sch).option("maxfilespertrigger",1).csv('/FileStore/tables/student_database/')
display(df)
```

```
df.writeStream.format('console').trigger(processingTime="10 seconds").outputMode('append').start()
```
