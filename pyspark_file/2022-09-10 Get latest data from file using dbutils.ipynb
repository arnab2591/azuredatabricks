```
# File location and type
file_location = "/FileStore/tables/employee.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
df.show()
```

```
df.repartition(2).write.format("parquet").partitionBy("Year").mode("overwrite").save("/user/hive/warehouse/Employee")
```

```
def get_latest_file(path):
    list_of_file = dbutils.fs.ls(path)
    latest_file =[int(i.name.split('Year=')[1][:-1]) for i in list_of_file if '_SUCCESS' not in i.name]
    return max(latest_file)
```

```
def get_latest_file(path):
    list_of_file = dbutils.fs.ls(path)
    latest_file =[(int(i.name.split('Year=')[1][:-1]),i.path) for i in list_of_file if '_SUCCESS' not in i.name]
    return max(latest_file)
```

```
path = get_latest_file('/user/hive/warehouse/Employee/')[1]
print(path)
df = spark.read.format("parquet").load(path)
df.show()
```
