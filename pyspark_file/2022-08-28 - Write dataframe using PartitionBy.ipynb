```
simpleData = [(1,"Sagar","CSE","UP",80),(2,"Shivam","IT","MP",86),(3,"Muni","Mech","AP",70)]

columns = ["ID","StudentName","Department","City","Marks"]
df_1 = spark.createDataFrame(data = simpleData,schema = columns)
df_1.show()
```

```
simpleData1 = [(1,"Sagar","CSE","UP",80),(3,"Muni","Mech","AP",70)]

columns = ["ID","StudentName","Department","City","Marks"]
df_2 = spark.createDataFrame(data = simpleData1,schema = columns)
df_2.show()
```

```
df=df_1.union(df_2)
df.show()
```

```
df.write.option("header","true").mode("overwrite").csv("/FileStore/tables_output/Sample_Partition_Output")
df.show()
```
