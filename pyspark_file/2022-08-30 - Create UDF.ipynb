```
simpleData = [(1,"Sagar","CSE","UP",80),(2,"Shivam","IT","MP",86),(3,"Muni","Mech","AP",70)]

columns = ["ID","StudentName","Department","City","Marks"]
df_1 = spark.createDataFrame(data = simpleData,schema = columns)
df_1.show()
```

```
simpleData1 = [(5,"Raj","CSE","HP"),(7,"Kunal","Mech","Rajasthan")]

columns = ["ID","StudentName","Department","City"]
df_2 = spark.createDataFrame(data = simpleData1,schema = columns)
df_2.show()
```

```
from pyspark.sql.functions import lit
# fill up missing value using lit command 
df_2 = df_2.withColumn("Marks",lit("null"))
df = df_1.union(df_2)
df.show()
```

```
input = "SaGar"
output = "sAgAr"
```

```
def convertcase(st):
    c=""
    for i in st:
        if ('a'<=i<='z'):
            c=c+i.upper()
        else:
            c=c+i.lower()
    return c
```

```
a = convertcase(input)
print(a)
```

```
from pyspark.sql.functions import col,udf
convertcase_df = udf(convertcase)
df.select(col("ID"),convertcase_df(col("StudentName")).alias("StudentName")).show()
```

```
df.select(col("ID"),convertcase_df(col("StudentName")).alias("StudentName"),convertcase_df(col("City")).alias("City")).show()
```

```
# Create a view or table with DataFrame

temp_table_name = "Sample_csv"

df.createOrReplaceTempView(temp_table_name)
```
