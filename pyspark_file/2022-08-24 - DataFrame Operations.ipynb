```
simpleData = [("Sagar","CSE","UP",80),("Shivam","IT","MP",86)]

columns = ["StudentName","Department","City","Marks"]
df_1 = spark.createDataFrame(data = simpleData,schema = columns)
df_1.show()
```

```
simpleData = [("Sagar","CSE","UP",80),("Muni","Mech","AP",70)]

columns = ["StudentName","Department","City","Marks"]
df_2 = spark.createDataFrame(data = simpleData,schema = columns)
df_2.show()
```

```
df_1.union(df_2).show()
```

```
df_1.union(df_2).distinct().show()
```

```
df_3 = df_2.union(df_1).distinct()
df_3.show()
```

```
# 1st method of When and Otherwise statement
from pyspark.sql.functions import col,when
df_4 = df_3.withColumn("State",when(col("City")=="UP","Uttar Pradesh").when(col("City")=="AP","Andhra Pradesh").when(col("City")=="MP","Madhya Pradesh").otherwise("Unknown"))
df_4.show()
```

```
# 2nd method of When and Otherwise statement
from pyspark.sql.functions import col,when
df_4 = df_3.select(col("*"),when(col("City")=="UP","Uttar Pradesh").when(col("City")=="AP","Andhra Pradesh").when(col("City")=="MP","Madhya Pradesh").otherwise("Unknown").alias("State"))
df_4.show()
```

```
from pyspark.sql.functions import col,when
df_4 = df_3.select(col("StudentName"),when(col("City")=="UP","Uttar Pradesh").when(col("City")=="AP","Andhra Pradesh").when(col("City")=="MP","Madhya Pradesh").otherwise("Unknown").alias("State"))
df_4.show()
```
