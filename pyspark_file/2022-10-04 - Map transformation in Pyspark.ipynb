```
df = spark.read.option('header',True).option('inferschema',True).csv('/FileStore/tables/Fillna.csv')
df.show()
```

```
df1 = df.rdd.map(lambda x:(x[0]*100+1,x[1],x[2]))
df1.collect()
```

```
df1 = df.na.fill(value=0,subset=["ID"])
df1 = df1.rdd.map(lambda x:(x[0]*100+1,x[1],x[2]))
df1.collect()
```

```
df2 = df1.toDF(["ID","Name","Age"])
df2.show()
```

```
def funmap(x):
    return x.ID*100,x.Name,x.Age
```

```
df1 = df.na.fill(value=0,subset=["ID"])
df1 = df1.rdd.map(lambda x:funmap(x))
df1.collect()
```

```
df2 = df1.toDF(["ID","Name","Age"])
df2.show()
```
