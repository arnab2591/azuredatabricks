```
list = [("Banana",1000,"USA"),("Carrots",1500,"USA"),("Beans",1600,"USA"),("Orange",2000,"USA"),("Orange",2000,"USA"),
        ("Banana",400,"China"),("Carrots",1200,"China"),("Beans",1500,"China"),("Orange",4000,"China"),("Banana",2000,"Canada"),
       ("Carrots",2000,"Canada"),("Beans",2000,"Mexico")]

df = spark.createDataFrame(list,schema=('Fruits','Price','Country'))
df.show()
```

```
df1 = df.groupBy('Fruits').sum('Price')
df1.show()
```

```
df1 = df.groupBy('Fruits').pivot('Fruits').sum('Price')
df1.show()
```

```
df2 = df.groupBy('Fruits','Country').sum('Price')
df2.show()
```

```
df2 = df.groupBy('Fruits').pivot('Country').sum('Price')
df2.show()
```
