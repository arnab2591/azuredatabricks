```
simpleData = [(1,"Sagar","CSE","UP",80),(2,"Shivam","IT","MP",86),(3,"Muni","Mech","AP",70)]

columns = ["ID","StudentName","Department","City","Marks"]
df_1 = spark.createDataFrame(data = simpleData,schema = columns)
df_1.show()
```

```
simpleData1 = [(1,"Sagar","CSE","UP",80),(3,"Muni","Mech","AP",70)]

columns = ["ID","StudentName","Department","City","Marks"]
df_2 = spark.createDataFrame(data = simpleData1,schema = columns)
df_2.show()
```

```
df_1.join(df_2,df_1.ID==df_2.ID,"inner").show()
```

```
df_1.join(df_2,df_1.ID==df_2.ID,"left").show()
```

```
from pyspark.sql.functions import col
df_1.alias("A").join(df_2.alias("B"),col("A.ID")==col("B.ID"),"inner").show()
```

```
df_1.alias("A").join(df_2.alias("B"),col("A.ID")==col("B.ID"),"inner").select(col("A.ID").alias("ID_A"),col("B.StudentName"),col("B.Department")).show()
```

```
df2=df_1.union(df_2)
df2.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number,rank
window = Window.partitionBy("ID").orderBy("Marks")
df2.withColumn("r1",row_number().over(window)).show()
```

```
window = Window.partitionBy().orderBy("Marks")
df2.withColumn("r1",row_number().over(window)).show()
```

```
window = Window.partitionBy().orderBy(col("Marks").desc())
df2.withColumn("r1",row_number().over(window)).show()
```

```
window = Window.partitionBy(["ID","StudentName","City"]).orderBy(col("Marks").desc())
df2.withColumn("r1",row_number().over(window)).select(col("ID"),col("StudentName"),col("City"),col("r1")).show()
```

```
window = Window.partitionBy(["ID","StudentName","City"]).orderBy(col("Marks").desc())
df2.withColumn("r1",rank().over(window)).select(col("ID"),col("StudentName"),col("City"),col("r1")).show()
```
